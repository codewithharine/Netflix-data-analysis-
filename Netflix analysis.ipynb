```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/adamerose/datasets/master/netflix_titles.csv"
df = pd.read_csv(url)
df.head()
```

```python
# 4. Check Missing Values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# 5. Movies vs TV Shows
sns.countplot(x='type', data=df, palette='pastel')
plt.title('Count of Content Type')
plt.xlabel('Type')
plt.ylabel('Total')
plt.show()
```

```python
# 6. Top 10 Genres
df['listed_in'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genres")
plt.show()
```

```python
# 7. Content Added Over Years

# Fix date formatting issues (stripping spaces, handling errors)
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce', infer_datetime_format=True)

# Extract year from date
df['year_added'] = df['date_added'].dt.year

# Plot content added by year
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
